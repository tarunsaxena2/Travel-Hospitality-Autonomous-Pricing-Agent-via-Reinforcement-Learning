# Week 1 — MDP & Environment Design

## 1. Environment Setup & Imports

```python
import gymnasium as gym
import numpy as np

print("All imports successful")
```

## 2. State Space Definition

```python
# State = (remaining_inventory, days_until_departure)
observation_space = gym.spaces.Box(
    low=np.array([0, 0]),
    high=np.array([100, 30]),
    dtype=np.float32
)
print("Observation space:", observation_space)
```

## 3. Action Space Definition

```python
# Action = discrete price level bin (e.g. 10 price tiers)
action_space = gym.spaces.Discrete(10)
print("Action space:", action_space)
```

## 4. Reward Function

```python
def reward_function(price_level, units_sold):
    """Reward = revenue generated in this timestep."""
    return price_level * units_sold

# Validate with sample values
sample_cases = [
    (50, 3),
    (80, 1),
    (20, 5),
    (100, 0),
]

for price, units in sample_cases:
    r = reward_function(price, units)
    print(f"price={price}, units_sold={units} -> reward={r}")
```

## MDP Formulation (Formal)

- **State (S):** (remaining_inventory, days_until_departure)
- **Action (A):** discrete price_level bin selected by the agent
- **Transition (P):** inventory -= units_sold; days_until_departure -= 1
- **Reward (R):** price_level * units_sold
- **Horizon (T):** episode ends when days_until_departure = 0 or inventory = 0
