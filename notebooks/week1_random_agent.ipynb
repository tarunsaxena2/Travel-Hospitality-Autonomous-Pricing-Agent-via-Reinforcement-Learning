# Week 1 — Random Agent Baseline

## 1. Setup

```python
import sys
sys.path.append('../src')
import gymnasium as gym
from random_agent import RandomAgent

print("Imports successful")
```

## 2. Random Agent Definition

```python
# See src/random_agent.py for the RandomAgent class implementation
# Agent samples uniformly from the valid action space at every step
print("RandomAgent class loaded from src/random_agent.py")
```

## 3. Test Run

```python
# Placeholder env until Member 1's PricingEnv is merged
test_env = gym.make("CartPole-v1")
agent = RandomAgent(test_env.action_space)

obs, info = test_env.reset()
print("Initial observation:", obs)

for step in range(5):
    action = agent.act(obs)
    obs, reward, terminated, truncated, info = test_env.step(action)
    print(f"Step {step}: action={action}, reward={reward}, done={terminated or truncated}")
    if terminated or truncated:
        break

test_env.close()
print("Test run complete — RandomAgent works correctly with Gymnasium API")
```

## 4. Notes on Gymnasium API

- `gym.Env` requires: `__init__`, `reset()`, `step(action)`, `render()`
- `observation_space` / `action_space` define valid states and actions
- MDP = (S, A, P, R, γ): state, action, transition, reward, discount factor

## 5. Run 500 Episodes — Record Episodic Revenue

```python
episode_rewards = []

for episode in range(500):
    obs, info = test_env.reset()
    total_reward = 0
    done = False

    while not done:
        action = agent.act(obs)
        obs, reward, terminated, truncated, info = test_env.step(action)
        total_reward += reward
        done = terminated or truncated

    episode_rewards.append(total_reward)

print("Completed", len(episode_rewards), "episodes")
print("First 10 episode rewards:", episode_rewards[:10])
```

## 6. Revenue Distribution — Histogram & Summary Statistics

```python
import numpy as np
import matplotlib.pyplot as plt

mean_revenue = np.mean(episode_rewards)
std_revenue = np.std(episode_rewards)

print(f"Mean episodic revenue: {mean_revenue:.2f}")
print(f"Std dev episodic revenue: {std_revenue:.2f}")

plt.figure(figsize=(8, 5))
plt.hist(episode_rewards, bins=30, edgecolor='black')
plt.axvline(mean_revenue, color='red', linestyle='dashed', linewidth=1.5, label=f'Mean = {mean_revenue:.2f}')
plt.title("RandomAgent — Episodic Revenue Distribution (500 episodes)")
plt.xlabel("Episodic Revenue")
plt.ylabel("Frequency")
plt.legend()
plt.tight_layout()
plt.savefig('../outputs/random_agent_revenue_histogram.png', dpi=150)
plt.show()
```
